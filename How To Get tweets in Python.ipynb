We will use 03 libraries tweepy, configparser and pandas

```python
import tweepy
import configparser
import pandas as pd
```

```python
#read configs
config = configparser.ConfigParser()
config.read('config.ini')
```

```python
api_key = config ['twitter']['api_key']
api_key_secret=config['twitter']['api_key_secret']

access_token =config['twitter']['access_token']
access_token_secret=config['twitter']['access_token_secret']
```

```python
#authentication

auth =tweepy.OAuthHandler(api_key, api_key_secret)
auth.set_access_token(access_token,access_token_secret)

api=tweepy.API(auth)
```

```python
public_tweets = api.home_timeline()
print(public_tweets )
```

```python
# To see the 1st tweet
print(public_tweets[0].text)
```

```python
# To see all the tweets:
for tweet in public_tweets: 
    print(tweet.text)
```

```python
# To see the tweet time :
print(public_tweets[0].user.created_at)

# To see who tweeted this one:
print(public_tweets[0].user.screen_name)
```

```python
# We will save these data into csv file using pandas
```

```python
columns=['Time', "User", 'Tweet']
data=[]

for tweet in public_tweets:
    data.append([tweet.created_at, tweet.user.screen_name, tweet.text ])

df=pd.DataFrame(data, columns= columns)
print(df)
```

```python
# Let Saving now into csv file
df.to_csv('tweets.csv')
```
